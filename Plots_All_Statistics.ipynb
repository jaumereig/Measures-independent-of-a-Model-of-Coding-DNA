```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%%bash
# FASTA to tbl format for AWK programs
./FastaToTbl fasta/HUMHBB.fa > fasta/HUMHBB.tbl
```

# Measures based on periodic correlations between nucleotide positions
- Position Assymetry
- Periodic Assymetry Index
- Average Mutual Information
- Fourier Spectrum

#### Running in AWK

```python
%%bash
# run PA in awk
(time gawk -f ../Assymetry/Assymetry.awk 120 10 fasta/HUMHBB.tbl > output/Assymetry_awk_120_10.out) 2> output/time_Assymetry_awk_120_10.out
# run PAI in awk
(time gawk -f ../Correlation/PAI/PAI.awk 120 10 fasta/HUMHBB.tbl > output/PAI_awk_120_10.out) 2> output/time_PAI_awk_120_10.out
# run AMI in awk
(time gawk -f ../Correlation/AMI/AMI.awk 120 10 fasta/HUMHBB.tbl > output/AMI_awk_120_10.out) 2> output/time_AMI_awk_120_10.out
```

#### Running in Python

```python
%%bash
# run PA in python
(time python3 Assymetry.py 120 10 fasta/HUMHBB.fa > output/Assymetry_pyt_120_10.out) 2> output/time_Assymetry_pyt_120_10.out
# run PAI in python
(time python3 Periodic_Assymetric_Index.py 120 10 fasta/HUMHBB.fa > output/PAI_pyt_120_10.out) 2> output/time_PAI_pyt_120_10.out
# run AMI in python
(time python3 Average_Mutual_Information.py 120 10 fasta/HUMHBB.fa > output/AMI_pyt_120_10.out) 2> output/time_AMI_pyt_120_10.out
```

#### Position Assymetry

```python
# PLOT AWK OUTPUT
with open("output/Assymetry_awk_120_10.out", "r") as data:
    lines=data.readlines()
    nucl_assym_start_awk=[]
    nucl_assym_end_awk=[]
    value_assym_awk=[]
    counter=0
    for line in lines:
        nucl_assym_start_awk.append(float(line.split()[0]))
        nucl_assym_end_awk.append(float(line.split()[1]))
        value_assym_awk.append(float(line.split()[2]))
    counter+=1
plt.plot(nucl_assym_start_awk, value_assym_awk)
plt.show()
#plt.savefig("output/Assymetry_awk_120_10.png", dpi=300)

# PLOT PYTHON OUTPUT
with open("output/Assymetry_pyt_120_10.out", "r") as data:
    lines=data.readlines()
    nucl_assym_start_pyt=[]
    nucl_assym_end_pyt=[]
    value_assym_pyt=[]
    counter=0
    for line in lines:
        nucl_assym_start_pyt.append(float(line.split()[0]))
        nucl_assym_end_pyt.append(float(line.split()[1]))
        value_assym_pyt.append(float(line.split()[2]))
    counter+=1
plt.plot(nucl_assym_start_pyt, value_assym_pyt)
plt.show()
#plt.savefig("output/Assymetry_pyt_120_10.png", dpi=300)

# both plots together
plt.plot(nucl_assym_start_pyt, value_assym_pyt, '.')
plt.plot(nucl_assym_start_awk, value_assym_awk)
plt.show()
#plt.savefig("output/Assymetry_merged_120_10.png", dpi=300)
```

#### Fourier Spectrum

#### Execution time of each statistic
